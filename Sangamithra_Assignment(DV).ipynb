#### 1.	Import the attached Indian_food.csv file in Jupyter notebook and perform following data cleaning operations through operations using Pandas and visualizations using Matplotlib, Seaborn and Plotly:

```python
import numpy as np
import pandas as pd
```

```python
pwd
```

```python
df=pd.read_csv(r"E:\\Software Python\\Assignment Ques\\Data Visualisation\\Indian_food.csv")
df.head(10)
```

##### a)	In the state column, replace the -1 with the mode of state column.

```python
df['state'] = df['state'].astype(str)
mode = df[df['state'] != '-1']['state'].mode()[0]
df['state'] = df['state'].replace('-1', mode)
df.head(20)
```

##### b)	Whatever the state you got in previous question, fill the corresponding region in the region column.

```python
df['region'] = df['region'].astype(str)
mode = df[df['region'] != '-1']['region'].mode()[0]
df['region'] = df['region'].replace('-1', mode)
df.head(20)
```

##### c)	Replace the -1 with mean for cook time variable.

```python
df["cook_time"] = pd.to_numeric(df["cook_time"], errors="coerce")
mean = df[df["cook_time"] != -1]["cook_time"].mean()
df["cook_time"] = df["cook_time"].replace(-1, round(mean))
df.head(20)
```

##### d)	Most preferred flavours by number of customers using pie chart.

```python
import matplotlib.pyplot as plt
```

```python
flavors = df["flavor_profile"].value_counts(dropna=True)
plt.figure(figsize=(5,5))
plt.pie(flavors, labels=flavors.index, autopct="%1.1f%%", startangle=140)
plt.show()
```

##### e)	Most requested course by customers using Bar chart. Print the labels also on the top of each bar. Then sort the bar chart in descending order.

```python
counts = df["course"].value_counts().sort_values(ascending=False)
bars = plt.bar(counts.index, counts.values)
plt.bar_label(bars, labels=counts.values)
plt.xlabel("Course")
plt.ylabel("Count")
plt.show()
```

##### f)	Comparison between Vegetarians and Non-Veg orders in each region. Hint: You can use stacked/grouped bar chart for comparison. 

```python
counts = df.groupby(["region", "Diet"]).size().unstack(fill_value=0)
counts.plot(kind="bar")
plt.title("Comparison of Vegetarian and Non-Vegetarian Dishes by Region")
plt.xlabel("Region")
plt.ylabel("No. of Orders")
plt.show()
```

##### g)	Create the box plot for Order cooking time to see the distribution. Also check if there are any outliers in this column.

```python
plt.boxplot(df["cook_time"],vert=0)
plt.show()
```

Pie Chart

```python
x=['A', 'B', 'C', 'D']
y=np.random.randint(10, 100, size=4)
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```
